```python
n = int(input())
lst = []
for i in range(n):
    lst.append(int(input()))
lst.sort(reverse=True)
lst2 = [lst[i:i + 3] for i in range(0, len(lst), 3)]
pay = 0
for i in range(len(lst2)):
    if len(lst2[i]) == 3:
        pay += sum(lst2[i]) - min(lst2[i])
    else:
        pay += sum(lst2[i])
print(pay)
```
