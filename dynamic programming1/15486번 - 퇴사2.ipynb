```python
n = int(input())
t, p = [],[]
dp = [0] * (n+1)
for i in range(n):
    l=list(map(int, input().split()))
    t.append(l[0])
    p.append(l[1])
```

```python
for i in range(n):
    if t[i] <= n-i:
        dp[i+t[i]] = max(dp[i+t[i]], dp[i]+p[i])
    dp[i+1] = max(dp[i], dp[i+1])
print(dp[n])
```
