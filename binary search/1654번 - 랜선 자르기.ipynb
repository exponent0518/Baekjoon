```python
k, n = map(int, input().split())
l = [int(input()) for _ in range(k)]
start, end = 1, max(l) # 이분탐색 처음과 끝 위치
while start <= end:
    mid = (start + end) // 2 # 중간 위치
    lines = 0 # 랜선 수
    for i in l:
        lines += i // mid # 분할된 랜선 수
    
    if lines >= n:
        start = mid + 1
    else:
        end = mid - 1
print(end)
```
