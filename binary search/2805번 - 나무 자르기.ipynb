```python
n, m = map(int, input().split())
l = list(map(int, input().split()))
start, end = 1, max(l)
while start <= end:
    mid = (start+end) // 2
    spl = 0
    for i in l:
        if i >= mid:
            spl += i - mid
    
    #  
    if spl >= m:
        start = mid + 1
    else:
        end = mid - 1
print(end)
```
