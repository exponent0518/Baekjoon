```python
n=list(str(input()))
i=0
start=0

while i < len(n):
    if n[i] == '<':
        i+=1
        while n[i] != '>':
            i+=1
        i+=1
    elif n[i] == ' ':
        i+=1
    else:
        start = i
        while i < len(n) and n[i].isalnum():
            i+=1
        tmp = n[start:i] 
        tmp.reverse()       
        n[start:i] = tmp
print(''.join(n))
```
